The general rules of an NFL overtime (during the regular season) are as follows:

**Overtime rules for NFL Regular Season** (see  [here](https://operations.nfl.com/the-rules/nfl-overtime-rules/) for more details):

- The team that wins the coin toss gets to try first.
- If the first team (Team 1) scores a "Touchdown" on their first drive, they win and the game is over.  
- If the first team (Team 1) scores a "Field Goal" on their first drive, the other team (Team 2) gets a turn and the following holds: 
   - If Team 2 scores a Touchdown on their drive they win and the game is over.  
   - If Team 2 scores a Field Goal on their drive, Team 1 gets the ball back and the next score (before time runs out) wins the game.    
   - If Team 2 does not score on their drive before the time runs out, they lose and the game is over.  
- If the first team (Team 1) does not score on their first drive, Team 2 gets the ball and the next score (before time runs out) wins the game.  
- The game ends at a tie if neither team scores at all within 10 minutes OR if they each scored Field Goals in their first drives and then no scores followed in an overall duration of 10 minutes.
